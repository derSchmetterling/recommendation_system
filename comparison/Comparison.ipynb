<h2> Content-based Filtering Recomendation </h2>

Here, I create another dataframe dropping the 'id' axis, since this column it`s not important for the cosine comparison.

At last, I will sum the columns of the playlist_comparison dataframe to have a vector that represents the entire playlist.

```python
playlist_comparison = playlist_comparison.drop('id', axis = 1)
searchspace_comparison = searchspace_comparison.drop('id', axis = 1)
```

```python
playlist_vector = playlist_comparison.sum(axis = 0)
playlist_vector
```

<h2> Finding smimilarity </h2>

The cosine similarity it's just the cosine between two vectors. In our case, it will measure the cosine between the playlist_vector and every row in our searchspace_comparison dataframe. After that, I will just sort 40 of the entries with higher similarity and our recommendation system will be ready. 

$$ Cosine Sim(A,B) = \frac{A . B}{||A||||B||} \frac{\sum_{i=1}^{n}A_i B_i}{\sqrt{\sum_{i=1}^{n}A_i^2}\sqrt{\sum_{i=1}^{n}B_i^2}}$$

```python

def generate_playlist_recos(non_playlist_df, features, nonplaylist_features):
    '''
    Generated recommendation based on songs in aspecific playlist.
    ---
    Input: 
    df (pandas dataframe): spotify dataframe
    features (pandas series): summarized playlist feature (single vector)
    nonplaylist_features (pandas dataframe): feature set of songs that are not in the selected playlist
        
    Output: 
    non_playlist_df_top_40: Top 40 recommendations for that playlist
    '''
    
    # Find cosine similarity between the playlist and the complete song set
    non_playlist_df['sim'] = cosine_similarity(nonplaylist_features.values, features.values.reshape(1, -1))[:,0]
    non_playlist_df_top_40 = non_playlist_df.sort_values('sim',ascending = False).head(40)
    
    return non_playlist_df_top_40

```

```python
results_reco = generate_playlist_recos(df, playlist_vector, searchspace_comparison)
results_reco
```
